```python
from sklearn.model_selection import train_test_split # tree
from sklearn.neighbors import KNeighborsClassifier  #neighbor theory
from sklearn import linear_model
```

```python
iris_data=iris["data"]
iris_target=iris["target"]
train_data,test_data,train_ans,test_ans=train_test_split(iris_data,iris_target,test_size=0.3) # 整數=比數  小數=百分比 30% test用
```

```python
KNN=KNeighborsClassifier()
KNN.fit(train_data,train_ans)
KNN.predict(test_data)
```

```python
test_ans
```

```python
KNN.score(test_data,test_ans)
```

```python
line=linear_model.LinearRegression() # 線性回歸
```

```python
line.fit(train_data,train_ans)
line.predict(test_data)
```

```python
line.score(test_data,test_ans)
```
