```python
import pandas as pd
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris.keys()
```

```python
x=pd.DataFrame(iris["data"],columns=iris["feature_names"])
y=pd.DataFrame(iris["target"],columns=["target_name"])
```

```python
x
```

```python
y
```

```python
df=pd.DataFrame.join(x,y)
```

```python
df
```

```python
data=pd.concat([x,y],axis=1) # 1=horizon 0=verticaL
data
```

```python
data.describe()
```

```python
data.sort_values(by="sepal length (cm)",ascending=False)
```

```python
data[data["petal length (cm)"]<3]
```

```python
data.groupby("target_name").mean()
```
