```python
df.groupby("縣市")["確定病例數"].sum() #以縣市作為群組 把裡面的病例數作加總
```

```python
df.groupby("年齡層")["確定病例數"].max() #min(),count()
```

```python
df.groupby("年齡層")["確定病例數"].sum()
```

```python
df.groupby("縣市")["確定病例數"].agg(["sum","max","min","count","mean"])
```
